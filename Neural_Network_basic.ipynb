## [V.3](https://youtu.be/tMrbN67U9d4?list=PLQVvvaa0QuDcjD5BAw2DxE6OF2tius3V3) Dot Products

Using loops

```python
inputs = [1,2,3,2.5]

weights =   [
            [0.2,0.8,-0.5,1.0],
            [0.5, -0.91, 0.26, -0.5],
            [-0.26, -0.27, 0.17, 0.87],
            ]

biases = [2,3,0.5]

layer_outputs = [] # output of the calculated layer

for neuron_weights, neuron_bias in zip(weights, biases): # zip combines two list into a list where each of the nth element is a list of the nth element of orogianl lists
    neuron_output = 0
    for a_input, weight in zip(inputs,neuron_weights):
        neuron_output += a_input*weight
    neuron_output += neuron_bias
    layer_outputs.append(neuron_output)

print("Layer output: ", layer_outputs)
```

Using **.dot()** of **Numpy** on **nD Array **

First dot param defines the dimension of the output (dim of 1st_in = dim of out) 

so sequence matters

```python
inputs = [1,2,3,2.5] # shape (4)

weights =   [  #shape (3,4)
            [0.2,0.8,-0.5,1.0],
            [0.5, -0.91, 0.26, -0.5],
            [-0.26, -0.27, 0.17, 0.87],
            ]

biases = [2,3,0.5]

# output = np.dot(weights, inputs) + biases
output = np.dot(inputs, np.array(weights).T) + biases
print(output)
```

Dealing with objects

```python
np.random.seed(0)

X = [
    [1.0, 2.0, 3.0, 2.5],
    [2.0, 5.0, -1.0, 2.0],
    [-1.5, 2.7, 3.3, -0.8]
    ]

class Layer_Dense:
    def __init__(self, n_inputs, n_neurons): # constructor of class
        self.weights = np.random.randn(n_inputs, n_neurons) * 0.1
        self.biases = np.zeros((1, n_neurons))
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases

layer1 = Layer_Dense(4,5)
layer2 = Layer_Dense(5,2)

layer1.forward(X)
layer2.forward(layer1.output)
print(layer2.output)
```

## [V.5](https://youtu.be/gmjzbpSVY1A?list=PLQVvvaa0QuDcjD5BAw2DxE6OF2tius3V3) Hidden layer activation functions

1. Rectified linear (ReLU)

3. Linear
4. Softmax

These functions comes after we complete our multiplication of **Weights** and addition of **biases**

###step function AKA unit step function



**Pros:**
1. very simple
2. Fast to calculate

**Cons:**
1. not very useful for correcting values of weights and bases in backward propagqation

###Sigmoid Function



**Pros:**
1. gives granularity in terms of training the network

**Cons:**
1. Slow in calculation 
2. for very large or small values gives diminishing result 

###Rectified linear Function



**Pros:**
1. doest have the problem of both sigmoid and step function
2. fast to calculate

###Genetare Dataset

```python
np.random.seed(0)

def spiral_data(samples, classes): #data generator
    points = samples
    X = np.zeros((points*classes, 2))
    y = np.zeros(points*classes, dtype='uint8')
    for class_number in range(classes):
        ix = range(points*class_number, points*(class_number+1))
        r = np.linspace(0.0, 1, points)  # radius
        t = np.linspace(class_number*4, (class_number+1)*4, points) + np.random.randn(points)*0.2
        X[ix] = np.c_[r*np.sin(t*2.5), r*np.cos(t*2.5)]
        y[ix] = class_number
    return X, y

import matplotlib.pyplot as plt

X, y = spiral_data(100, 3)

plt.scatter(X[:,0], X[:,1])
plt.show()
print("")
plt.scatter(X[:,0], X[:,1], c=y, cmap="brg")
plt.show()
```

## [V.6](https://youtu.be/omz_NdFgWyU?list=PLQVvvaa0QuDcjD5BAw2DxE6OF2tius3V3) Softmax activation function for output layer 



```python
import math

layer_outputs = [4.8, 1.21, 2.385]

E = 2.71828182846 
#or
E = math.e

exp_values = []

for output in layer_outputs:
    exp_values.append(E**output)

print(exp_values)

norm_base = sum(exp_values)
norm_values = []

for values in exp_values:
    norm_values.append(values / norm_base)

print(norm_values)
print(sum(norm_values))
```

```python
# !pip install nnfs
import nnfs

nnfs.init()
# np.random.seed(0)


class Layer_Dense:
    def __init__(self, n_inputs, n_neurons): # constructor of class
        self.weights = np.random.randn(n_inputs, n_neurons) * 0.1
        self.biases = np.zeros((1, n_neurons))
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases

class Activation_ReLU:
    def forward(self, inputs):
        self.output = np.maximum(0, inputs)

class Activation_Softmax:
    def forward(self, inputs):
        exp_values = np.exp(inputs - np.max(inputs, axis=1, keepdims=True))
        probabilities = exp_values / np.sum(exp_values, axis=1, keepdims=True)
        self.output = probabilities 

X, y = spiral_data(samples=100, classes=3)

dense1 = Layer_Dense(2,3) # 2 features of each input 
activation1 = Activation_ReLU()

dense2 = Layer_Dense(3,3) # 3 for output from previous layer and 3 classes of data
activation2 = Activation_Softmax()

dense1.forward(X)
activation1.forward(dense1.output)

dense2.forward(activation1.output)
activation2.forward(dense2.output)

print(activation2.output[:5]) #first 5 of 300 samples 
```
